## Question 1: Comparing Two Lists Using a Line Plot

Problem Statement:
You are given two lists representing monthly sales of two products over 6 months.
You need to visualize how their sales trends compare over time using a line plot.

Approach:
1. Create two lists: months and sales.
2. Use `sns.lineplot()` to plot both lines on the same graph.
3. Add appropriate labels and a title to make the visualization clear.

```python
# Write your code here
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example lists
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [45, 55, 60, 70, 65, 75]
sales_B = [40, 50, 58, 68, 70, 80]

sns.lineplot(x=months, y=sales_A, label='Sales A')
sns.lineplot(x=months, y=sales_B, label='Sales B')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison for Sales A and Sales B')
plt.show()
```

## Question 4: Relationship Between Distance and Fare

Problem Statement:
Does the distance of a taxi trip affect the fare charged?

## Question 5: Average Fare by Day of the Week

Problem Statement:
Do taxi fares vary across different days of the week?

Approach:
1. Use the 'pickup_day' or derive day information from the datetime column if available.
2. Plot the average fare per day using `sns.barplot()`.

```python
taxis = sns.load_dataset('taxis')

if 'pickup_day' not in taxis.columns and 'pickup' in taxis.columns:
    taxis['pickup_day'] = taxis['pickup'].dt.day_name()

avg_fare_per_day = taxis.groupby('pickup_day')['fare'].mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).reset_index()

sns.barplot(x='pickup_day', y='fare', data=avg_fare_per_day)

plt.xlabel('Day of the Week')
plt.ylabel('Average Fare Amount ($)')
plt.title('Average Taxi Fare by Day of the Week')
plt.show()
```

## Question 8: Trip Duration vs. Fare Amount

Problem Statement:
Is there any relationship between how long a trip takes and how much it costs?

Approach:
1. Use columns like 'pickup' and 'dropoff' to calculate trip duration (if timestamps are available).
2. Plot a scatter plot between duration and fare using `sns.scatterplot()`.

```python
taxis = sns.load_dataset('taxis')
if 'pickup' in taxis.columns and 'dropoff' in taxis.columns:
    taxis['trip_duration'] = (taxis['dropoff'] - taxis['pickup']).dt.total_seconds() / 60

sns.scatterplot(x='trip_duration', y='fare', data=taxis)
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Fare Amount ($)')
plt.title('Relationship Between Trip Duration and Fare Amount')
plt.show()
```

## Question 10: Average Tip by Distance Category

Problem Statement:
We want to know if longer trips generally result in higher tips.

Approach:
1. Create distance categories (e.g., short, medium, long) using the 'distance' column.
2. Use `sns.barplot()` to plot average tips for each distance category.

```python
taxis = sns.load_dataset('taxis')
bins = [0, 2, 5, taxis['distance'].max()]
labels = ['Short', 'Medium', 'Long']
taxis['distance_category'] = pd.cut(taxis['distance'], bins=bins, labels=labels, include_lowest=True)
sns.barplot(x='distance_category', y='tip', data=taxis)

plt.xlabel('Distance Category')
plt.ylabel('Average Tip Amount ($)')
plt.title('Average Tips by Taxi Trip Distance Category')

plt.show()
```
